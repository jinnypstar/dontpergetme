#### 3. 데이터 탐색하기
##### 3.1 상위 10개 회사별 향수 보유 현황 시각화

```python
# 필요한 라이브러리 import
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 데이터프레임 불러오기
df = pd.read_csv('../Data/preprocess-data/final_perfume-info.csv')

# Dior 과 Christian Dior 회사 합치기 (같은 회사이므로)
df['company'] = df['company'].replace(['Dior', 'Christian Dior'], 'Dior')

# 회사별 향수 개수 세기
company_counts_updated = df['company'].value_counts().head(10)

# 회사명 개수로 수직 그래프 그리기
plt.figure(figsize=(14, 8))
sns.barplot(y=company_counts_updated.values, x=company_counts_updated.index, palette='mako')
plt.title('Top 10 Companies with the Most Perfumes')
plt.ylabel('Number of Perfumes')
plt.xlabel('Company')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.show()
```
